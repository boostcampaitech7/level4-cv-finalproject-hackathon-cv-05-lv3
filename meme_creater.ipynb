```python
import csv
import re

def parse_meme_data(raw_data):
    meme_data = []
    current_year = None
    
    for line in raw_data.splitlines():
        year_match = re.match(r"==== (\d{4})년 ====", line)
        if year_match:
            current_year = int(year_match.group(1))
            continue
        
        month_match = re.match(r"\s*-\s*(\d{1,2})월:\s*(.+)", line)
        if month_match and current_year:
            month = month_match.group(1)
            memes = re.findall(r'\[\[([^\]|]+)(?:\|[^\]]+)?\]\]', month_match.group(2))
            for meme in memes:
                meme_data.append({"Year": current_year, "Month": f"{month}월", "Meme": meme.strip()})
    
    return meme_data

raw_data = """==== 2015년 ====

- 1월: [[sake L]]
- 3월: [[오로나민C]], [[Yee]]
- 6월: [[누가 기침 소리를 내었는가]]
- 7월: [[나 꿈꿨어 귀신 꿈꿨어]]
- 9월: [[조태오|어이가 없네]]
- 10월: [[레스토랑스]]

==== 2016년 ====

- 2월: [[상어 가족|아기상어]]
- 4월: [[CHEER UP|샤샤샤]]
- 5월: [[곡성|뭣이 중헌디]]
- 6월: [[너 때문에 흥이 다 깨져버렸으니까 책임져]]
- 10월: [[너굴맨]]
- 11월: [[내가 이러려고 대통령을 했나]]

==== 2017년 ====

- 8월: [[우원재|산타벌스]]
- 9월: [[트로피카나 스파클링]]
- 10월: [[Getting Over It with Bennett Foddy|항아리 게임]]
- 11월: [[상상도 못한 정체]]

==== 2018년 ====

- 1월: [[사랑을 했다]], [[폰은정]]
- 2월: [[였던 것 드립]], [[김영미|영미]]
- 4월: [[탈모르파티]], [[어벤져스: 인피니티 워]]
- 5월: [[다 아는 사람들이구먼]]
- 6월: [[나는 예쁘지 않습니다(동영상)|나는 예쁘지 않습니다]]
- 7월: [[오또케]], [[He is ○○]]
- 10월:[[껍질 미리 깐 달걀]]
- 11월: [[소년점프]]
- 12월: [[핵인싸]]

==== 2019년 ====

- 2월: [[대충 드립]], [[자전차왕 엄복동]]
- 3월: [[안녕하살법]]
- 4월: [[MIROH#음원|힘들지 않아 거친 정글 속에 뛰어든 건 나니까 암오케]]
- 5월: [[나도 내가 징징거리고 눈꼴시려운건 알고 있는데]], [[쁘악수]]
- 6월: [[기생충(영화)|기생충]], [[Woman Yelling at a Cat]]
- 7월: [[이 시국에]], [[제리 마우스#인사 짤|인사하는 제리]]
- 8월: [[이이잉 앗살라말라이쿰]], [[곽철용]], [[유튜브 알고리즘]], [[여름이었다]
- 9월: [[오뚝이 다트]], [[찌리찌리]], [[헬창]], [[국밥충]], [[어벤져스: 엔드게임]]
- 10월: [[가성비 댓글]], [[조커(2019)|코리안조커]]
- 11월: [[펭수]], [[아마존 익스프레스#아마존 환영 세레모니 밎 밈화]], [[겨울왕국 2]]
- 12월: [[단또단또]], [[야나두]]

==== 2020년====

- 1월: [[홍정오/아임뚜렛|아임뚜렛]], [[그랜절]], [[지이잉]], [[호박고구마]]
- 2월: [[아무노래]], [[코로나바이러스감염증-19]]
- 3월: [[종로로 갈까요]], [[만희물]], [[ㄷㄷㄷㅈ]], [[나비보벳따우]], [[미안하다 이거 보여주려고 어그로끌었다]]
- 4월: [[관짝춤]], [[나는 개인이오]], [[캬루/밈|캬루]]
- 5월: [[깡(노래)|깡]], [[헬테이커]], [[펀쿨섹좌]], [[엄준식(인터넷 밈)|엄준식]]
- 6월: [[~은(는) 사드세요.....제발]], [[불 좀 꺼줄래?]], [[야 꿀벌]], [[가면라이더 드립]]
- 7월: [[첵스 파맛]], [[군침이 싹 도노]], [[가짜사나이/1기|가짜사나이]], [[극장판 짱구는 못말려: 부리부리 왕국의 보물|부리부리 마신 소환 춤]]
- 8월: [[뒷광고]], [[테스형!]], [[애기공룡 둘리]], [[돌리랑 도트가 제일 좋아]], [[사실은 오래전부터 당신 같은 XXX를 기다려 왔다우]]
- 9월: [[제주도 찐 사투리]], [[어몽어스]], [[빅맥송]], [[다메다메]], [[펫 더 피포]]
- 10월: [[Team Azimkiya]], [[뭉탱이]], [[난 너가 줏대있게 인생 살았으면 좋겠어]]
- 11월: [[치즈분수]], [[한국인이 좋아하는 속도]]
- 12월: [[사쿠란보]], [[하츠네 형돈]], [[Pop Cat]], [[당연히 말이 되죠]]

==== 2021년 ====

- 1월: [[무야호]], [[팝캣]], [[강아지 경찰 아저씨]], [[포브스]] 드립, [[스윗중남]], [[비트 밈]]
- 2월:  [[삼각함수송]], [[극장판 귀멸의 칼날: 무한열차 편|무한열차]]
- 3월: 가수 [[일론 머스크]], [[시무라 아주머니]], [[LH 게이트]]
- 4월: [[학교폭력 멈춰!|멈춰!]][* 사회 문제 해결엔 별 도움이 안되고 그저 황당함에 웃음만 나온다는 점에서 밈이 되었다.], [[마이야히]], [[육군 we 육군]]
- 5월: [[머니게임(웹예능)|머니게임]], [[나락송]], [[아라아라]], [[프라이데이 나이트 펑킨]], [[삼성걸]], [[근데 이제 뭐함?]], [[Next Level(aespa)|넥스트 레벨]], [[굿바이 선언]]
- 6월: [[꽁기깅깅깅공강강꽁기깅깅꽁기깅강]], [[제로투 댄스]], [[한심좌]]
- 7월: [[2020 도쿄 올림픽/픽토그램 판토마임 공연|2020 도쿄 올림픽 픽토그램]]
- 8월: [[2020 도쿄 올림픽/양궁|활벤져스]], [[소금소금소금팥팥팥]], [[잭 오#잭 오 챌린지]], [[그런건 없다 게이야]], [[성윤모(신병)|성윤모]]
- 9월: [[오징어 게임]], [[앙카 존 댄스]], [[라고 할 뻔]]
- 10월: [[스트릿 우먼 파이터]], [[설거지론]], [[커다란 자갈돌은 짱돌]], [[파피 플레이타임]]
- 11월:  [[몰?루]], [[북방코끼리바다표범(밈)|북방코끼리바다표범]]
- 12월:  [[어느새부터 힙합은 안 멋져]], [[힘숨찐]], [[Super idol的笑容|슈퍼 아이돌]], [[사실이 아닙니다!]], [[뽀구미|예쁘잖아]]

==== 2022년====

- 1월: [[피에로 드립]], [[미스터 인크레더블 밈]]
- 2월: [[저 반바지 아니에요]]
- 3월: [[포켓몬빵]], [[킹근육골드몬]], [[지구방위대 챌린지]], [[어몽어스 잼민이]]
- 4월: [[부럽지가 않어]], [[마이 네임 이즈 찢기]], [[LOVE DIVE]]
- 5월: [[마미 롱 레그]], [[소울리스좌]], [[That That]], [[마키노 츠쿠시#밈|아이 러브 마이 화자 마자 브라자]], [[꼬마펭귄 핑구|눗눗]], [[범죄도시 2]]
- 6월: [[재즈를 뭐라고 생각하세요?]], [[후드에서 살아남기]], [[여기어때송]], [[오붕가]]
- 7월: [[이상한 변호사 우영우]], [[닭팔이]], [[모기송(만다린)|모기송]], [[World's Smallest Violin|점점 빨라지는 노래]]
- 8월: [[제네시스좌]], [[슬픈 고양이 댄스|슬픈 고양이]], [[인주 앨리스]], [[알빠노]]
- 9월: [[실패작 소녀]], [[Young한데? 완전 MZ인데요?]], [[새삥]], [[제로투 회피]]
- 10월: [[중요한 것은 꺾이지 않는 마음]], [[갱갱갱]]
- 11월: [[알빠임|알빠임?]], [[귀여워서 미안해]], [[짱구야 아빠를 속인거니?]]
- 12월: [[뉴진스의 하입보이요]], [[근데 어쩔건데?]], [[폼 미쳤다]]

==== 2023년 ====

- 1월: [[닛몰캐쉬]], [[팅팅송]], [[더 글로리|멋지다 연진아, 스튜어디스 혜정아]], [[제기랄#제기랄, 또 ~~~야!|제기랄, 또 ~~~야!]], [[뉴진스의 하입보이요]], [[우리 엄마엄마가]]
- 2월: [[Sigma(밈)|시그마]], [[마따끄]], [[나문희의 첫사랑]], [[나이트 댄서]]
- 3월: [[고도로 발달한 A는 B와 구분할 수 없다]], [[스즈메의 문단속]], [[인터넷 야메로]], [[폿핏포|뽀삐뽀삐뽀 뽀삐뽀]]
- 4월: [[Happy Happy Happy|해피캣]], [[이궈궈던]]
- 5월: [[Emet Sound#아디아디아디|아디아디아디]], [[강풍 올백]], [[전생했더니 슬라임이었던 건에 대하여|~한 건에 대하여]], [[【최애의 아이】(애니메이션 1기)|【최애의 아이】]]
- 6월: [[너 T야?]], [[AI 커버]], [[사카밤바스피스#s-3]], [[스키비디 토일렛]], [[사우스 코리안 파크]], [[그리메이스 셰이크]], [[퀸카]]
- 7월: [[크아아악! XX아!]], [[지구 온라인]],[[엘리멘탈]], [[Steal The Show]]
- 8월: [[LK-99|LK-99(초전도체)]], [[인생 하드모드]], [[Red Flags#모스부호 밈|모스부호]], [[어 맞아맞아 놀랍지만 그건 사실이야]], [[김채원|피어나 너 내 도도독]]
- 9월: [[올리버쌤 성대모사]], [[탕후루]], [[워싱시]], [[Love Lee]], [[Miku]]
- 10월: [[김도와 마즈피플]], [[고죠 사토루/밈|잘 가라, 최강. 내가 없는 시대에 태어났을 뿐인 범부여]], [[I am신뢰에요~]], [[똑똑한 청년.]]
- 11월: [[장충동왕족발보쌈]], [[I am신뢰에요~]],[[당근칼]], [[기습숭배]]
- 12월: [[나 XX인데 개추 눌렀다]],[[(XXX에서 연락옴)]], [[치피치피 차파차파]], [[전두광]][* "실패하면 XX, 성공하면 XX 아입니까!", "그 이왕이면 XX(이)란 멋진 단어를 쓰십시오!"]

==== 2024년 ====

- 1월: [[치피치피 차파차파]], [[꽁꽁 얼어붙은 한강 위로 고양이가 걸어다닙니다]], [[국가권력급]],  [[한잔해]], [[웡카|움파룸파춤]]
- 2월: [[캣냅]], [[브링방방]],, [[첫 만남은 계획대로 되지 않아]]
- 3월: [[밤양갱(음반)#밤양갱|밤양갱]], 구구단 댄스, [[래빗 홀]],[[원영적 사고]]
- 4월: [[잘자요 아가씨|야레야레 못말리는 아가씨]]
- 5월: [[Pedro Pedro Pedro]], [[원영적 사고|완전 럭키비키잖아]], [[평화누리특별자치도]], [[김채원|점심을 든든히 먹어두어라 저녁은 지옥에서 먹을테니]], [[마라탕후루(Malatanghulu)|마라탕후루]], 포철고 챌린지, [[경북대학교 에브리타임]], 귀여워ㅋㅋ 키스할래요?, [[정상화(인터넷 밈)|정상화]](팩트는 ~라는 거임), [[수수수수퍼노바]], [[푸른 산호초]]
- 6월 : [[오승환/밈#s-2.2|형이랑 내기할래?]], [[류정란]] 챌린지, [[사랑했나봐(윤도현)|사랑했나봐]][* 이 음원에 곰돌이 [[푸]]가 [[강남스타일]]을 추는 영상이 유행중이다.], [[이승엽/여담#s-3.3|츠~긍슴승 승르흐르르~]], [[지가○○노코노코○○탓탓|시카노코노코노코코시탄탄]], [[버튜버 페이셜 테스트]], [[티라미수 케익]], [[하루만 기다리면 ○○가 나와요!]], [[뉴진스럽다]], [[English or Spanish]], [[폴라레티]], [[메스머라이저]]
- 7월 : [[네 기꺼이]], [[외모 췤!]], [[아로나 댄스]], [[줄게]], [[정상화(인터넷 밈)|신창섭]], [[쿵쿵따(서이브)|쿵쿵따]], [[미룬이#s-4|미룬이 사건]], [[삐끼삐끼 춤]], [[인사이드 아웃 2]], [[김쁠뿡]], [[조이는 보이]]
- 8월 : [[창팝]]([[다 해줬잖아]], [[바로 리부트 정상화]], [[이해가쏙쏙되잖아리슝좍아]]), [[삐끼삐끼 춤]], [[신박천문연구소|천문학자 밈]]
- 9월 :  [[Nikocado Avocado|두 수 앞]], [[두들 댄스]], [[캐치! 티니핑 시리즈|티니핑]], 바니토사, [[흑백요리사: 요리 계급 전쟁(시즌 1)|흑백요리사: 요리 계급 전쟁 시즌 1]]([[최강록]], [[유비빔]], [[[https://youtu.be/8PvxTMeiVJU|고죠](https://youtu.be/8PvxTMeiVJU%7C%EA%B3%A0%EC%A3%A0) 백종원]]), [[주우후테이 라덴#s-5|마이타케]], [[제로 산소]], [[마루는 강쥐]]
- 10월 : [[주우후테이 라덴#s-5|마이타케]], [[두들 댄스]], [[흑백요리사: 요리 계급 전쟁(시즌 1)|흑백요리사: 요리 계급 전쟁 시즌 1]](검은 장갑, [[영역 전개]]), 옴브리뉴, [[자낳괴 탄생비화]], 반전 어둠 챌린지, [[조커: 폴리 아 되]], [[젠레스 존 제로/OST#s-3.5|버니스 댄스]], [[[https://youtube.com/shorts/ftgJjAsHl3k|프로토](https://youtube.com/shorts/ftgJjAsHl3k%7C%ED%94%84%EB%A1%9C%ED%86%A0) 디스코]], [[피크민]]
- 11월 : [[APT.|아파트]], [[터미널]] 댄스([[실루엣]] 챌린지), Faces, [[차가운 상어 아가씨]], [[Queen never cry]], [[까탈레나]], [[샹하이 로맨스]]
- 12월 : [[계엄]], [[봇이지 뭐]], [[샹하이 로맨스]], [[바람피면 D지는거야]], [[미무카와 나이스 트라이]], [[테토리스]], [[오징어 게임(시즌 2)|오징어 게임 2]](얼음!!, 저는 이 게임을 해봤어요!, 둥글게 둥글게, [[타노스(오징어 게임)|타노스]] 랩), [[안주거리]], [[Kanaria|BRAIN]], [[[https://youtu.be/AfPmtV3ho8M?si=QfEhR4UGR0q5FpDx|Không](https://youtu.be/AfPmtV3ho8M?si=QfEhR4UGR0q5FpDx%7CKh%C3%B4ng) Sao Cả]](괜찮아 챌린지)
==== 2025년 ====
- 1월: [[오징어 게임(시즌 2)|오징어 게임 2]](얼음!!, 저는 이 게임을 해봤어요!, 둥글게 둥글게, [[타노스(오징어 게임)|타노스]] 랩), [[기가채드|내면의 기가차드]], [[나쁜 답변 좋은 답변]], [[칠 가이]], [[테토리스]]]"""
result = parse_meme_data(raw_data)

# CSV 파일로 저장
with open('meme_data.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['Year', 'Month', 'Meme'])
    writer.writeheader()
    for row in result:
        writer.writerow(row)

print("CSV 파일이 생성되었습니다: meme_data.csv")
```
