```python
import pandas as pd
import re
import os
```

```python
# 현재 작업 디렉토리 확인
current_dir = os.getcwd()
# 상대 경로에서 파일 경로 설정
csv_path = os.path.join(current_dir, "..", "crawling_books_data.csv")
# CSV 파일 불러오기
df = pd.read_csv(csv_path)
df
```

```python
# '책 설명'이 존재하는 경우만 처리
df.loc[df["책 설명"].notna(), "책 설명"] = df["책 설명"].str.replace(r'도서안내', '', regex=True).str.strip()
df
```

```python
# '책 설명'이 존재하는 경우만 처리
if "책 설명" in df.columns:
    def extract_sections(text):
        if pd.isna(text):  # '책 설명'이 NaN이면 그대로 반환 (pass)
            return None, None, None, None

        # '사서의 추천 글' 추출
        match = re.search(r'(?:사서의 추천 글|사서의 추천)\s*(.*?)(?=\s*저자 소개|책 속 한 문장|책 속 한문장|함께 읽으면 좋은 책|$)', text, re.DOTALL)
        sa_seo_recommendation = match.group(1).strip() if match else None

        # '저자 소개' 추출
        match = re.search(r'저자 소개\s*(.*?)(?=\s*책 속 한 문장|책 속 한문장|함께 읽으면 좋은 책|$)', text, re.DOTALL)
        author_intro = match.group(1).strip() if match else None

        # '책 속 한 문장' 또는 '책 속 한문장' 추출
        match = re.search(r'(?:책 속 한 문장|책 속 한문장)\s*(.*?)(?=\s*함께 읽으면 좋은 책|$)', text, re.DOTALL)
        quote_from_book = match.group(1).strip() if match else None

        # '함께 읽으면 좋은 책' 추출
        match = re.search(r'함께 읽으면 좋은 책\s*(.*)', text, re.DOTALL)
        recommended_books = match.group(1).strip() if match else None

        return sa_seo_recommendation, author_intro, quote_from_book, recommended_books

    # 책 설명이 NaN이 아닌 경우만 처리
    mask = df["책 설명"].notna()
    extracted_data = df.loc[mask, "책 설명"].apply(extract_sections)

    # 개별 컬럼에 할당
    df.loc[mask, "사서의 추천 글"] = extracted_data.apply(lambda x: x[0] if x else None)
    df.loc[mask, "저자 소개"] = extracted_data.apply(lambda x: x[1] if x else None)
    df.loc[mask, "책 속 한 문장"] = extracted_data.apply(lambda x: x[2] if x else None)
    df.loc[mask, "함께 읽으면 좋은 책"] = extracted_data.apply(lambda x: x[3] if x else None)
```

```python
# '사서의 추천 글', '저자 소개', '책 속 한 문장', '함께 읽으면 좋은 책'이 모두 None인 행 찾기
columns_to_check = ["사서의 추천 글", "저자 소개", "책 속 한 문장", "함께 읽으면 좋은 책"]
rows_to_update = df[columns_to_check].isna().all(axis=1)

# 해당 행의 '책 설명' 값을 '사서의 추천 글'로 이동
df.loc[rows_to_update, "사서의 추천 글"] = df.loc[rows_to_update, "책 설명"]

# '책 설명' 컬럼 드랍
df.drop(columns=["책 설명"], inplace=True, errors="ignore")
df
```

```python
df.to_csv('processed_books_data.csv', index=False, encoding="utf-8-sig")
```
